```python
# Read in CSV
internet_access_df = pd.read_csv('../resources/regression_ml_model/usa_merged_internet_access.csv')
```

```python
# Convert Internet Score & Round Level Up
internet_access_df['internet_score'] = round(internet_access_df['level_up_access_pop_percent'] * 10,1)
internet_access_df['level_up_access_pop_percent'] = round(internet_access_df['level_up_access_pop_percent'],2)
internet_access_df.head()
```
